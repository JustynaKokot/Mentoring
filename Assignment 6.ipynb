```python
#Task 1f

for index, row in movie_performance.iterrows():
    if row['Rotten Tomatoes Critic Score'] >= 90:
        movie_performance.loc[index, 'Recommended'] = 'Strongly Yes'
    elif row['Rotten Tomatoes Critic Score'] >= 60:
        movie_performance.loc[index, 'Recommended'] = 'Yes'
    elif row['Rotten Tomatoes Critic Score'] >= 40:
        movie_performance.loc[index, 'Recommended'] = 'No'
    else:
        movie_performance.loc[index, 'Recommended'] = 'Strongly No'
        
movie_performance
```
