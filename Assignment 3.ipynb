```python
#Zadanie 5 - przekształcenie zadania 4 w słownik

vowels = ['a', 'ą', 'e', 'ę', 'i', 'o', 'u', 'ó', 'y',]

def number_of_vovwels(word): 
    lower = word.lower()
    lower.split()
    letters = list(lower)
    slownik = {}
    for letter in letters:
        if letter in vowels:
            if letter in slownik:
                slownik[letter] = slownik[letter] + 1
            else:
                slownik[letter] = 1
    print(slownik)
    
number_of_vovwels('Borys')
```

```python
#Zadanie 7 - alfabet Morsa

def Morse_code_translator(x):
    word = x.lower()
    new_list = []
    for letter in word:
        if letter == 'a':
            a = '.-'
            new_list.append(a)
        elif letter == 'b':
            b = '-...'
            new_list.append(b)
        elif letter == 'c':
            c = '-.-.'
            new_list.append(c)
        elif letter == 'd':
            d = '-..'
            new_list.append(d)
        elif letter == 'e':
            e = '.'
            new_list.append(e)
        elif letter == 'f':
            f = '..-.'
            new_list.append(f)
        elif letter == 'g':
            g = '--.'
            new_listappend(g)
        elif letter == 'h':
            h = '....'
            new_list.append(h)
        elif letter == 'i':
            i = '..'
            new_list.append(i)
        elif letter == 'j':
            j = '.---'
            new_list.append(j)
        elif letter == 'k':
            k = '-.-'
            new_list.append(k)
        elif letter == 'l':
            l = '.-..'
            new_list.append(l)
        elif letter == 'm':
            m = '--'
            new_list.append(m)
        elif letter == 'n':
            n = '-.'
            new_list.append(n)
        elif letter == 'o':
            o = '---'
            new_list.append(o)
        elif letter == 'p':
            p = '.--.'
            new_list.append(p)
        elif letter == 'q':
            q = '--.-'
            new_list.append(q)
        elif letter == 'r':
            r = '.-.'
            new_list.append(r)
        elif letter == 's':
            s = '...'
            new_list.append(s)
        elif letter == 't':
            t = '-'
            new_list.append(t)
        elif letter == 'u':
            u = '..-'
            new_list.append(u)
        elif letter == 'v':
            v = '...-'
            new_list.append(v)
        elif letter == 'w':
            w = '.--'    
            new_list.append(w)
        elif letter == 'x':
            x = '-..-'
            new_list.append(x)
        elif letter == 'y':
            y = '-.--'
            new_list.append(y)
        elif letter == 'z':
            z = '--..'
            new_list.append(z)
        elif letter == ' ':
            space = '/'
            new_list.append(space)
        elif letter == '.':
            dot = '.-.-.-'
            new_list.append(dot)

    #print(new_list)
    print("".join(new_list))

Morse_code_translator('kot')
```
